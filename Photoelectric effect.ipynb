Einsteins Photoelectric effect

1) Photoelectric effect on copper plate

```python
import numpy as np
import matplotlib.pyplot as plt


h = 6.626e-34  # Planck's constant (J·s)
e = 1.602e-19  # Charge of an electron (C)
phi = 4.53 * e  # Work function of copper (J) (4.5 eV)

frequencies = np.linspace(-1e13, 5e15, 500)  # Adjust frequency range (Hz) to start above threshold frequency


photon_energy = h * frequencies

kinetic_energy = np.maximum(photon_energy - phi, 0)



plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 2)
plt.plot(frequencies, kinetic_energy / e, label='Kinetic Energy (eV)', color='blue')
plt.axvline(phi / h, color='red', linestyle='--', label='Threshold Frequency (Copper)')
plt.title("Kinetic Energy vs Frequency", fontsize=14)
plt.xlabel("Frequency (Hz)", fontsize=12)
plt.ylabel("Kinetic Energy (eV)", fontsize=12)
plt.xlim(frequencies[0], frequencies[-1])  
plt.ylim(-5, np.max(kinetic_energy) / e + 0.2)  
plt.axhline(-5, color='black', linewidth=0.5) 
plt.legend()
plt.grid(True)

plt.show()

max_kinetic_energy = np.max(kinetic_energy)
stopping_potential = max_kinetic_energy / e  # In volts

print(f"Threshold Frequency for Copper: {phi / h:.2e} Hz")
print(f"Maximum Kinetic Energy of Ejected Electrons: {max_kinetic_energy / e:.2f} eV")
print(f"Stopping Potential: {stopping_potential:.2f} V")
```

2) Photoelectric effect on Iron|

```python
import numpy as np
import matplotlib.pyplot as plt


h = 6.626e-34  # Planck's constant (J·s)
e = 1.602e-19  # Charge of an electron (C)
phi = 4.73 * e  # Work function of copper (J) (4.5 eV)


frequencies = np.linspace(-1e13, 5e15, 500)  # Adjust frequency range (Hz) to start above threshold frequency
intensity = 1.0  # Light intensity (arbitrary units)

photon_energy = h * frequencies


kinetic_energy = np.maximum(photon_energy - phi, 0)  # KE = hν - φ, but KE >= 0

num_electrons = np.where(photon_energy > phi, intensity * (photon_energy - phi) / photon_energy, 0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 2)
plt.plot(frequencies, kinetic_energy / e, label='Kinetic Energy (eV)', color='blue')
plt.axvline(phi / h, color='red', linestyle='--', label='Threshold Frequency (Copper)')
plt.title("Kinetic Energy vs Frequency", fontsize=14)
plt.xlabel("Frequency (Hz)", fontsize=12)
plt.ylabel("Kinetic Energy (eV)", fontsize=12)
plt.xlim(frequencies[0], frequencies[-1])  
plt.ylim(-0.2, np.max(kinetic_energy) / e + 0.2)  
plt.axhline(0, color='black', linewidth=0.5) 
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()


max_kinetic_energy = np.max(kinetic_energy)
stopping_potential = max_kinetic_energy / e  # In volts

print(f"Threshold Frequency for Iron: {phi / h:.2e} Hz")
print(f"Maximum Kinetic Energy of Ejected Electrons: {max_kinetic_energy / e:.2f} eV")
print(f"Stopping Potential: {stopping_potential:.2f} V")
```
