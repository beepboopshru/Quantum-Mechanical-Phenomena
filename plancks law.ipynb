Planck's Law for Black Body Radiation

```python
import numpy as np
import matplotlib.pyplot as plt


h = 6.626e-34       # Planck's constant (J·s)
c = 3.0e8           # Speed of light (m/s)
k_B = 1.381e-23     # Boltzmann constant (J/K)

def energy_density_wavelength(wavelength, temperature):
    numerator = 8 * np.pi * h * c / wavelength**5
    exponent = h * c / (wavelength * k_B * temperature)
    denominator = np.exp(exponent) - 1
    return numerator / denominator


wavelengths = np.linspace(1e-7, 3e-6, 1000)  # From 100 nm to 3000 nm (UV to infrared)

temperatures = [1000 , 3000,5400, 4000, 5000, 6000 ]  # Temperatures in Kelvin

plt.figure(figsize=(10, 6))
for T in temperatures:
    density = energy_density_wavelength(wavelengths, T)
    plt.plot(wavelengths * 1e9, density, label=f"T = {T} K")  # Convert meters to nanometers for plot


plt.title("Spectral Energy Density of Blackbody Radiation (Wavelength)", fontsize=16)
plt.xlabel("Wavelength (nm)", fontsize=14)
plt.ylabel("Energy Density ($J/m^3/m$)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()


plt.show()
```
